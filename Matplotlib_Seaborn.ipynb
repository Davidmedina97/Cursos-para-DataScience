#Colores visualizados

```python
import matplotlib.colors as mcolors
import matplotlib.patches as mpatch

overlap = {name for name in mcolors.CSS4_COLORS
           if f'xkcd:{name}' in mcolors.XKCD_COLORS}

fig = plt.figure(figsize=[9, 5])
ax = fig.add_axes([0, 0, 1, 1])

n_groups = 3
n_rows = len(overlap) // n_groups + 1

for j, color_name in enumerate(sorted(overlap)):
    css4 = mcolors.CSS4_COLORS[color_name]
    xkcd = mcolors.XKCD_COLORS[f'xkcd:{color_name}'].upper()

    # Pick text colour based on perceived luminance.
    rgba = mcolors.to_rgba_array([css4, xkcd])
    luma = 0.299 * rgba[:, 0] + 0.587 * rgba[:, 1] + 0.114 * rgba[:, 2]
    css4_text_color = 'k' if luma[0] > 0.5 else 'w'
    xkcd_text_color = 'k' if luma[1] > 0.5 else 'w'

    col_shift = (j // n_rows) * 3
    y_pos = j % n_rows
    text_args = dict(fontsize=10, weight='bold' if css4 == xkcd else None)
    ax.add_patch(mpatch.Rectangle((0 + col_shift, y_pos), 1, 1, color=css4))
    ax.add_patch(mpatch.Rectangle((1 + col_shift, y_pos), 1, 1, color=xkcd))
    ax.text(0.5 + col_shift, y_pos + .7, css4,
            color=css4_text_color, ha='center', **text_args)
    ax.text(1.5 + col_shift, y_pos + .7, xkcd,
            color=xkcd_text_color, ha='center', **text_args)
    ax.text(2 + col_shift, y_pos + .7, f'  {color_name}', **text_args)

for g in range(n_groups):
    ax.hlines(range(n_rows), 3*g, 3*g + 2.8, color='0.7', linewidth=1)
    ax.text(0.5 + 3*g, -0.3, 'X11/CSS4', ha='center')
    ax.text(1.5 + 3*g, -0.3, 'xkcd', ha='center')

ax.set_xlim(0, 3 * n_groups)
ax.set_ylim(n_rows, -1)
ax.axis('off')

plt.show()
```

#Grafica de barras

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population =[1000,800,900,1000,300]
#Set style
plt.style.use('grayscale')
#Set bar_plot
plt.barh(country,population,color=['aqua','aquamarine','green','azure'])
plt.yticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
#Set context
plt.title('World Population Country')
plt.xlabel('Country')
plt.ylabel('Population',rotation=0)
#Show plot
plt.show()
```

#Graficos de distribuciones

```python
sns.histplot(data = tip, x= 'tip', bins = 15, cumulative=True)
```

```python
sns.histplot(data = tip, x= 'tip', bins = 15, hue= 'sex', stat = 'density', multiple= 'dodge');
plt.show()
```

```python
sns.kdeplot(data=tip, x='tip', hue='sex', shade=True);
```

#Relaciones

```python
sns.scatterplot(data= tip, x= 'total_bill', y = 'tip', hue= 'day', style= 'time', size='size');
plt.legend(loc= 'center', bbox_to_anchor= (1.12, 0.5))
```

```python
markers = {"Lunch": 'D', "Dinner": 's'}
sns.scatterplot(data= tip, x= 'total_bill', y = 'tip', hue= 'day', style= 'time', size='size', markers = markers);
plt.legend(loc= 'center', bbox_to_anchor= (1.12, 0.5));
```

```python
sns.relplot(data= tip, x= 'total_bill', y = 'tip',  hue= 'day', style= 'time', size='size', kind= 'line', col = 'time');
```

#Jointplot y pairplot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist', marginal_ticks=True, #muestra una pequea tabla para la grfica exterior
marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge')); #marginal_kws se le pueden agregar diferentes parmetros
```

```python
sns.pairplot(data= tips);
```
